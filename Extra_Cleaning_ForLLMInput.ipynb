```python
# Step 1: Clean df (remove '[deleted]' and "nan" string)
df = df[df['author'] != '[deleted]']
df['body_text'] = df['body_text'].str.replace("nan", "", regex=False)

# Step 2: Drop 'media' column
if 'media' in df.columns:
    df = df.drop(columns=['media'])

# Step 3: Pull post_id from df2 and merge it into df
# Make sure both body_text columns are stripped and aligned
df['body_text'] = df['body_text'].str.strip()
df2['body_text'] = df2['body_text'].str.strip()

# Merge on both author and body_text (only keeps rows that match)
df = df.merge(df2[['author', 'body_text', 'post_id']], on=['author', 'body_text'], how='left')
```

```python
df8 = pd.read_csv('/Users/jacksonsorenson/Documents/Computational Media Lab/Weed Study/CSV for mass database organization/R:Dabs/NEW LLM REQDY/new_dab_llm_ready.csv')
```

```python
df8.head()
```

```python
# Step 1: Drop rows where body_text is NaN or deleted
df8 = df8[df8['body_text'].notna()]
df8 = df8[df8['body_text'].str.lower() != '[deleted]']

# Step 2: Clean body_text
def clean_text(text):
    text = text.strip()
    text = re.sub(r'\s+', ' ', text)  # normalize whitespace
    text = re.sub(r'[^\x00-\x7F]+', '', text)  # remove non-ASCII chars (emojis, etc.)
    return text

df8['clean_body_text'] = df8['body_text'].apply(clean_text)

# Step 3 (Optional): Truncate long posts (e.g., keep only first 500 characters)
df8['clean_body_text'] = df8['clean_body_text'].apply(lambda x: x[:500])

# Step 4: Drop rows where clean text is empty after cleaning
df8 = df8[df8['clean_body_text'].str.strip() != ""]

# Step 5: (Optional) Reset index
df8.reset_index(drop=True, inplace=True)

# View result
df8.head()
```

```python
# Save to CSV with only LLM-relevant columns
df8[['author', 'clean_body_text']].to_csv('/Users/jacksonsorenson/Desktop/cleaned_llm_ready.csv', index=False)
```
